## load the book

```python
with open("miracle.txt", "r", encoding='utf8', errors='ignore') as file: 
    book = file.read()
```

### how many chapters?

### using string methods

```python
book.count("Chapter")
```

### with regex

```python
import re
```

```python
pattern = re.compile("Chapter [0-9]+")
chapters = re.findall(pattern, book)
print(chapters)
len(chapters)
```

### find "love"

```python
pattern = re.compile("love")
love = re.findall(pattern, book)
print(love)
```

### find words as a list

```python
pattern = re.compile("[A-za-z]+")
findings = re.findall(pattern, book.lower())
len(findings)
findings[:5]
```

```python
d = {}
for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1 
    else:
        d[word] = 1
```
